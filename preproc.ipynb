```python
import pandas as pd
```

```python
data = pd.read_csv('zahedan-ghi-temp.csv')
DF = pd.DataFrame(data=data)
DF = DF.drop('Year',axis=1)
DF=DF.drop('Minute',axis=1)
DF
```

```python
old = 0
new = 24
DFF = pd.DataFrame(columns=['Month','Day','GHI','Temperature','Solar Zenith Angle','Surface Albedo','DHI'])

for i in range(2,668):

    DFtest = DF.iloc[old:new,:]
    df_compressed = DFtest.agg(['sum'])//24
    DFF = pd.concat([DFF, df_compressed],ignore_index=True)
    old=new
    new=24*i
    
DF = DFF.drop('Hour',axis=1)
DF = DFF.drop('DNI',axis=1)
DF
```
